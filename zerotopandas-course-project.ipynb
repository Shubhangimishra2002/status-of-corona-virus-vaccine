#### Q1:  Plot a graph related to the columns and show in it: people who are vaccinated per hundred, people who are totally vaccinated per hundred,total vaccinations of people per hundred.  

````python
plt.hist(vaccine_df.total_vaccinations_per_hundred);
plt.hist(vaccine_df.people_vaccinated_per_hundred);
plt.hist(vaccine_df.people_fully_vaccinated_per_hundred);
plt.title('Vaccinations of people per Hundred')
plt.legend(['people_vaccinated_per_hundred', 'people_totally_vaccinated_per_hundred','total_vaccination_per_hundred']);
````

In the above graph it is shown that people_vaccinated_per_hundred are more than total_vaccination_per_hundred.
this is because many vaccines or doses which also include boosters, can be counted individually. This can be because the same person may receive more
than one dose of vaccine, which perfectly the number of doses per 100 people can be higher than 100.

#### Q2: Compare the total vaccinations of people and the people who are vaccinated with the first dose of vaccine?

````python
plt.title("Data Comparison")
plt.hist(vaccine_df.total_vaccinations);
plt.hist(vaccine_df.people_vaccinated);
plt.legend(['total_vaccination','people_who_are_vaccinated'])
````

The above data shows that total vaccinations of people is slightly more than people who are vaccinated with the first dose of vaccine.
At least one dose of vaccination has been received by almost 63.5% of the total world population.
